```python
from skimage.io import imread
```

```python
image = imread('scikit-image.png')

# Let's check the type of this image data
print("Type:", type(image))

# See the shape of it
print("Shape:", image.shape)
```

```python
import matplotlib.pyplot as plt

plt.imshow(image)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(14, 28))
colors = {0: 'Red channel', 1: 'Green channel', 2:'Blue channel'}
for i in range(0, 3):
    # Get the corresponding color channel, last index is color index (channel)
    image_channel = image[:, :, i]
    print(colors[i], "Shape:", image_channel.shape)
    fig.add_subplot(1, 3, i+1)
    # Plot the channel image and add title
    plt.imshow(image_channel, cmap='gray')
    plt.title(colors[i])
plt.show()
```

```python
from skimage.transform import resize

# Resized image
resized_im = resize(image, output_shape=(70, 250))

fig=plt.figure(figsize=(25, 50))

# Plot original image
fig.add_subplot(1, 2, 1)
plt.imshow(image)
plt.title("Original Image with size: " + str(image.shape))

# Plot resized image
fig.add_subplot(1, 2, 2)
plt.imshow(resized_im)
plt.title("Resized Image with size: " + str(resized_im.shape))

plt.show()
```
